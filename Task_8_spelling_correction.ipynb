##Sentence spelling correction

```python
from textblob import TextBlob
```

```python
with open("sentence.txt", "r") as f:# Opening the text file
    text = f.read()  
    print(text)                   
    textBlb = TextBlob(text)            # Making a textblob
    textCorrected = textBlb.correct()   # Correcting the text
    print(textCorrected)
```

```python
with open("corrected_sentence.txt", "w") as f1:
    f1.write(str(textCorrected))
```

```python
#Function to compare two texts and seggregate
def compare(text1, text2):  
    l1 = text1.split()
    l2 = text2.split()
    good = 0
    bad = 0
    for i in range(0, len(l1)):
        if l1[i] != l2[i]:
            bad += 1
        else:
            good += 1
    return (good, bad)

# Helper function to calculate the percentage of misspelled words
def percentageOfBad(x):
    return (x[1] / (x[0] + x[1])) * 100
```

```python
with open("sentence.txt", "r") as f1: #typo-filled text
    t1 = f1.read()

with open("sentence_original.txt", "r") as f2: # contains the original text
    t2 = f2.read()

t3 = TextBlob(t1).correct()

mistakesCompOriginal = compare(t1, t2)
originalCompCorrected = compare(t2, t3)

print("Mistakes compared to original ", mistakesCompOriginal)
print("Original compared to corrected ", originalCompCorrected)

print("Percentage of mistakes in the test text: ", percentageOfBad(mistakesCompOriginal), "%")
print("Percentage of mistakes in the corrected text: ", percentageOfBad(originalCompCorrected), "%")
```

Here we can see that the mistakes in the text reduced from 21% to 12%. But here there are some limitations as the dataset used here is a manually created and typo filled one. With an original piece of text with spelling mistakes, the error rate will be considerably reduced.
